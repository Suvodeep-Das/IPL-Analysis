```python
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
pd = pd.read_csv('Top_100_bowlers.csv')
pd.head()
```

```python
#Top 10 Bowlers with most Number of wickets in the IPL .
highest_wickets = pd.nlargest(10,'Wkts')

plt.xticks(rotation=90)
sns.barplot(x='PLAYER',y='Wkts',data=highest_wickets)
sns.lineplot(x='PLAYER',y='Wkts',data=highest_wickets)
plt.xlabel('Player Name')
plt.ylabel('No. of wickets')
```

```python
#Top 10 Bowlers with best economy rate in the IPL .
economy = pd.sort_values(by = 'Econ', ascending = True).head(10)

plt.xticks(rotation = 90)
sns.swarmplot(x='PLAYER',y='Econ',data = economy)
sns.lineplot(x='PLAYER',y='Econ',data = economy)
plt.xlabel('Player Name')
plt.ylabel('Economy Rate')
```

```python
#Top 10 Bowlers with best Strike rate in the IPL .
bowler_sr = pd.sort_values(by = 'SR').head(10)

plt.xticks(rotation = 90)
sns.lineplot(x=bowler_sr.PLAYER, y = bowler_sr.SR)
sns.barplot(x=bowler_sr.PLAYER, y = bowler_sr.SR)
plt.xlabel('Player Name')
plt.ylabel('Strike Rate')
```

```python
#Top 10 Bowlers with best Bowling Average in the IPL 
bowler_avg = pd.sort_values(by = 'Avg').head(10)


plt.xticks(rotation=90)
sns.barplot(bowler_avg.PLAYER,bowler_avg.Avg)
sns.lineplot(bowler_avg.PLAYER,bowler_avg.Avg)
plt.xlabel('Player Name')
plt.ylabel('Average')
```

```python
#Top 10 Bowlers with most 4 wicket haul in the IPL .
wickets_4 = pd.nlargest(10,'4w')

plt.xticks(rotation=90)
sns.pointplot(x='PLAYER', y='4w',data = wickets_4)
sns.barplot(x='PLAYER', y='4w',data = wickets_4)
plt.xlabel('Player Name')
plt.ylabel('No. of 4 wicket haul')
```

```python
#Top 10 Bowlers with most 5 wicket haul in the IPL 
wickets_5 = pd.nlargest(10,'5w')

plt.xticks(rotation=90)
sns.barplot(x='PLAYER', y='5w',data = wickets_5)
sns.pointplot(x='PLAYER', y='5w',data = wickets_5)
plt.xlabel('Player Name')
plt.ylabel('No. of 5 wikcet haul')
```
